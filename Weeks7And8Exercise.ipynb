## Import xlsx File - BOING-BOING-CANDY-HIERARCHY-2016-SURVEY-Responses.xlsx

```python
# First I will Import xlsx File - BOING-BOING-CANDY-HIERARCHY-2016-SURVEY-Responses.xlsx
Candy_Responses_Data = pd.read_excel("BOING-BOING-CANDY-HIERARCHY-2016-SURVEY-Responses.xlsx")
Candy_Responses_Data.head()
```

```python
# Next I will use list and .columns function with dataframe object to 
# pull a list of my dataframes headers. 
list(Candy_Responses_Data.columns)
```

```python
# I will use pandas .rename function to edit my headers names. 
Candy_Responses_Data=Candy_Responses_Data.rename(columns={'Are you going actually going trick or treating yourself?':'Trick_or_Treating', 
                                  'Your gender:':'Gender',
                                  'How old are you?':'Age','Which country do you live in?':'Country',
                                                    'Which state, province, county do you live in?':'State',
                                                    '[Butterfinger]':'Butterfinger'})
Candy_Responses_Data.head()
```

## Reshape DataFrame

```python
# I will reshape the data by selecting only the data that has individuals that are 
# trick or treating to see what their opinions are for each candy
Trick_or_Treaters_Data = Candy_Responses_Data[Candy_Responses_Data['Trick_or_Treating'] == 'Yes']
Trick_or_Treaters_Data.head()
```

## Fill in Missing Data

```python
# First I will use the .notnull function to see if any of the columns 
# are missing data or not
Candy_Responses_Data.notnull()
# As seen below we see that we are missing some data.
```

```python
# Being that we are missing data and I do not want to just drop the 
# rows with missing data I have decided to use the .fillna function 
# to ffill and bfill the data which meanse if there is nothing above 
# we will back fill the data but if there is data in front we will 
# front fill the missing data
Candy_Responses_Data = Candy_Responses_Data.fillna(method='ffill').fillna(method='bfill')
Candy_Responses_Data
```

```python
# Next I will use .dropna to get rid of any other NaN that are 
# remaining that have not been taken care of
```

## Pivot DataFrame

```python
# Now that the missing data is filled we can create a pivot that the 
# indext is on the timestamp while the columns will be for Trick_or_Treating 
# and we will determine the age for our values which will tell us how old the 
# tricker treaters are.
Pivot_Candy_SURVEY_Data = Candy_Responses_Data.pivot(index='Timestamp', columns='Trick_or_Treating', values='Age')
Pivot_Candy_SURVEY_Data.head()
# As seen Below we have NaN due to the fact that either the individual 
# said yes or no to the fact that they went tricker treating 
```

## Grouping with Dicts/Series

```python
# First I will create a new DataFrame that shows just the Gender and Age
Updated_Candy_Responses_Data = Candy_Responses_Data.filter(['Gender','Age'], axis=1)
Updated_Candy_Responses_Data
```

```python
# Next I will use the .groupby function to bring together the Pivot_Candy_SURVEY_Data
# and show
Updated_Candy_Responses_Data.groupby(['Gender']).mean()
# As we can see the average age for a female is 38.34 and the average 
# age of a male is 42.23
```

## Crosstab

```python
# I will a simple cross-tabulation of two or more factors from our 
# dataframe through the use of .crosstab
pd.crosstab(Updated_Candy_Responses_Data['Age'], Updated_Candy_Responses_Data['Gender'])
# Seen below is the amount of males, females, others, and I'd rather 
# not say genders that are for each age group
```

## Generate date range

```python
# first I will ensure that the timestamp column is in date format
Candy_Responses_Data['Timestamp'] = pd.to_datetime(Candy_Responses_Data['Timestamp'])
```

```python
# next I will create a start date and end date to filter the data 
# frame with
start_date = '2016-10-24 05:14:24.625'
end_date = '2016-10-29 16:53:52.516'
```

```python
# Next I will create a mask that will be applied to the dataframe to 
# filter it
Candy_Mask = (Candy_Responses_Data['Timestamp'] > start_date) & (Candy_Responses_Data['Timestamp'] <= end_date)
```

```python
Masked_Candy_Data = Candy_Responses_Data.loc[Candy_Mask]
Masked_Candy_Data
# As seen below we have removed several lines from this data frame by 
# excluding dates that did not fall within the time frame
```
